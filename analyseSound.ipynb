```python
import sounddevice as sd
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf

# Set the audio parameters
duration = 10  # Duration of audio capture in seconds
sample_rate = 44100  # Sample rate in Hz
channels = 1  # Number of channels (mono)

# Capture audio from the microphone
print("Recording...")
recording = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=channels)
sd.wait()  # Wait until the recording is finished
print("Recording finished.")

# Convert the recorded audio to a numpy array
audio = np.squeeze(recording)

# Save the recorded audio as a WAV file
output_file = "recording_600Hz.wav"
sf.write(output_file, audio, sample_rate)
print(f"Recording saved as {output_file}")

# Calculate the decibel level of the recorded sound
reference_value = 1.0  # Reference value for decibel calculation (adjust as needed)
audio_db = 20 * np.log10(np.abs(audio) / reference_value)
average_db = np.mean(audio_db)

print(f"Average decibel level: {average_db:.2f} dB")

# Perform FFT on the audio data
fft_data = np.fft.fft(audio)
frequencies = np.fft.fftfreq(len(audio), 1 / sample_rate)

# Plot the FFT results
plt.figure(figsize=(8, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_data[:len(fft_data)//2]))
plt.xlabel('Frequency (Hz)')
plt.xlim(0,1000)
plt.ylabel('Magnitude')
plt.title('FFT of Recorded Audio')
plt.grid()
plt.show()
```
